### y2 va V2 be tartib ertefa va sorat ab zire dariche ast:

### $V_2 = V_1 - 2\sqrt{g*y_1} + 2\sqrt{g*y_2}$    (A)

### va hamchenin darim:

### $v_2 * y_2 = q_2$    (B)

## ja gozari rabete B dar A va hale an:


```python
from sympy.solvers import solve
from sympy import Symbol
y2 = Symbol("y")
a = V1 - (2 * ((g * y1)**(0.5)))
y2 = solve(a*y2 - 2 + (2*y2*((g * y2)**0.5)), y2) #y2[V1-2(gy1^0.5)+2(gy2^0.5)]-2 = 0 
y2 = float(y2[0])
y2
```

```python
V2 = q2 / y2
V2
```
