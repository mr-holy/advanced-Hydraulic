```python
# mostatili
V1 = 0.65 #m/s
y1 = 1.4 #m
q1 = V1 * y1 #m^2 / s
# bala bordan dariche:
q2 = 3 * q1
#V and q moj ?
print('q1 = %f and q2 = %f'%(q1,q2))
# assumed : V1 = V2 =>
y2 = 3 * y1
print("y2 = %f"%y2)
g = 9.81 #m^2/s
```

# positive surge

```python
Vw = V1 + (g * ((y2 / (2*y1)) * (y2 + y1))**0.5)

if Vw < 0:
    Vw = V1 - (g * ((y2 / (2*y1)) * (y2 + y1))**0.5)
else:
    print('halate mosbat')
Cs = Vw - V1
print('sorat enteshar moj %f mibashad'%Cs)
y_moj = abs(y1 - y2)
q_moj = y_moj * Cs
print('debi enteshar moj %f mibashad'%q_moj)
```
