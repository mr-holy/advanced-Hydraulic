```python
#kanale ariz
# V and Q moj?
y1 = 20 #ft
g = 32.2 #ft^3 / s
y2 = 10 #ft
C = 100 #ft^0.5 / s
S0 = 0.0003
```

```python
import IPython.display as ID
b = ID.Image("moj_tasvir.jpeg")
ID.display(b)
```

## ba dashtan zarib C va S0 az rabete zir soraat ra mohasebe mikonim (kanal ariz R = y):
## $V = C(y*S_0)^{0.5}$

```python
V1 = C *((y1 * S0)** 0.5)
V2 = C *((y2 * S0)** 0.5)
print('v1 = %f and V2 = %f'%(V1, V2))
```

# speed of wave
# $V_w = \frac{V_1y_1 - V_2y_2}{y_1 - y_2}$

```python
Vw = ((V1 * y1) - (V2 * y2)) / (y1 - y2)
Vw
```

# Q of the flow
# $q_r = y_2 ( V_w - V_2) = y_1 ( V_w - V_1)$

```python
qr = y2 * (Vw - V2)
qr1 = y1 * (Vw - V1)
print ("q(r) = %f = %f"%(qr1, qr))
```
